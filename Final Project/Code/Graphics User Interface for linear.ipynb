```python
import tkinter as tk
import csv 

#Created a window
window = tk.Tk()
window.geometry('1200x600')
window.title('House Price Estimator')

#Created both the labels and entries
label = tk.Label(master=window, text="This Estimates the price of a house using Linear Regression Coefficients ",fg="white",bg="black",width=100,height=5, font = ('calibre',15,'bold'))
label.pack()

labe2 = tk.Label(master=window, text="Lot_sqft",fg="black",width=0,height=0, font = ('calibre',10,'bold'))
labe2.pack()
labe2.place(x=10, y=150)

entry2 = tk.Entry(fg="black", width=50)
entry2.pack()
entry2.place(x=180, y=150)

labe3 = tk.Label(master=window, text="Number of Baths",fg="black",width=0,height=0, font = ('calibre',10,'bold'))
labe3.pack()
labe3.place(x=10, y=200)

entry3 = tk.Entry(fg="black", width=50)
entry3.pack()
entry3.place(x=180, y=200)

labe4 = tk.Label(master=window, text="Sqft",fg="black",width=0,height=0, font = ('calibre',10,'bold'))
labe4.pack()
labe4.place(x=10, y=250)


entry4 = tk.Entry(fg="black", width=50)
entry4.pack()
entry4.place(x=180, y=250)

labe5 = tk.Label(master=window, text="Number_of_beds",fg="black",width=0,height=0, font = ('calibre',10,'bold'))
labe5.pack()
labe5.place(x=10, y=300)


entry5 = tk.Entry(fg="black", width=50)
entry5.pack()
entry5.place(x=180, y=300)

labe6 = tk.Label(master=window, text="livability_index",fg="black",width=0,height=0, font = ('calibre',10,'bold'))
labe6.pack()
labe6.place(x=10, y=350)


entry6 = tk.Entry(fg="black", width=50)
entry6.pack()
entry6.place(x=180, y=350)

# function to calculate coefficients


def converter():
    Lot_sqft = float(entry2.get()) * -3002
    Lot_sqft = float(entry3.get()) * 77541
    houwse_Sqft = float(entry4.get()) * 987872
    number_of_beds = float(entry5.get()) * -306738
    livability_index = float(entry6.get()) * 41363
    total_value_in_dollars = (Lot_sqft + Lot_sqft +houwse_Sqft + number_of_beds + livability_index + 1185934) / 3000
    labe4["text"] = total_value_in_dollars

    # Button to activate the function
button = tk.Button(master= window, text="Click me to find the price ",width=25, height=5,bg="blue",fg="yellow", command=converter )
button.pack()
button.place(x=500, y=400)

# Label to print the price
labe4 = tk.Label(master=window, fg="black",width=0,height=0, font = ('calibre',20,'bold'))
labe4.pack()
labe4.place(x=700, y=500)

window.mainloop()
```
