```python
import pandas as pd
import tkinter as tk
import csv 
from tkinter import scrolledtext
from tkinter.scrolledtext import *

# imported libraries

window = tk.Tk()
#window = tk.Toplevel()
window.geometry('1500x600')
window.title('House Price Estimator')

# Background Image
image1 = tk.PhotoImage(file = r'C:\Users\goshl\OneDrive\Documents\GitHub\house_PNG50.png')

# Imported Data

data = pd.read_csv(r"C:\Users\goshl\OneDrive\Documents\GitHub\Project-1\fintech-Project-2\Joanne\cleaned_data.csv")
data = data[['ZipCode','City', 'Beds', 'Baths' , 'Acres', 'Lot_sqft' , 'Price']]

# Created Labels ()

label6 = tk.Label(master=window, image=image1)
label6.place(x= 0 , y = 0)


labe1 = tk.Label(master=window, text="GROUP 3 Realtor",fg="black",width=0,height=0, font = ('calibre',25,'bold'))
labe1.pack()
labe1.place(x=10, y=10)

labe2 = tk.Label(master=window, text="start price",fg="black",width=0,height=0, font = ('calibre',15,'bold'))
labe2.pack()
labe2.place(x=10, y=150)

entry2 = tk.Entry(fg="black", width=25)
entry2.pack()
entry2.place(x=250, y=150)

labe3 = tk.Label(master=window, text="end price",fg="black",width=0,height=0, font = ('calibre',15,'bold'))
labe3.pack()
labe3.place(x=10, y=200)

entry3 = tk.Entry(fg="black", width=25)
entry3.pack()
entry3.place(x=250, y=200)

labe5 = tk.Label(master=window, text="Number of Bedrooms",fg="black",width=0,height=0, font = ('calibre',15,'bold'))
labe5.pack()
labe5.place(x=10, y=250)

entry5 = tk.Entry(fg="black", width=25)
entry5.pack()
entry5.place(x=250, y=250)


labe8 = tk.Label(master=window, text="Number of Bath",fg="black",width=0,height=0, font = ('calibre',15,'bold'))
labe8.pack()
labe8.place(x=10, y=300)

entry8 = tk.Entry(fg="black", width=25)
entry8.pack()
entry8.place(x=250, y=300)



labe9 = tk.Label(master=window, text="ZipeCode",fg="black",width=0,height=0, font = ('calibre',15,'bold'))
labe9.pack()
labe9.place(x=10, y=350)

entry9 = tk.Entry(fg="black", width=25)
entry9.pack()
entry9.place(x=250, y=350)

## Created function 

def converter():
    
    start = float(entry2.get()) 
    end = float(entry3.get())
    house_price = data.set_index('Price').sort_index()
    data_price_loc = data_price.loc[start:end]
    
    num_Bath_loc_entry = float(entry8.get())
    data_bedroom_loc = data_price_loc.set_index('Baths')
    num_Bath_loc = (pd.DataFrame(data_bedroom_loc.loc[num_Bath_loc_entry])).head(50)
    
    
    num_beds_loc = num_Bath_loc.set_index('Beds')
    num_beds_loc_entry = float(entry5.get())
    FINAL = (pd.DataFrame(num_beds_loc.loc[num_beds_loc_entry])).head(50)
    

    labe4.insert(tk.END, FINAL)

    
def price():
    house_price = data.set_index('Price').sort_index()
    start = float(entry2.get()) 
    end = float(entry3.get())
    house_price = (pd.DataFrame(house_price.loc[start:end])).head(50)
    labe4.insert(tk.END, house_price)
    
def num_Bath():
    num_Bath_loc = data.set_index('Baths')
    num_Bath_loc_entry = float(entry8.get())
    num_Bath_loc = (pd.DataFrame(num_Bath_loc.loc[num_Bath_loc_entry])).head(50)
    labe4.insert(tk.END, num_Bath_loc)
    
def num_beds():
    num_beds_loc = data.set_index('Beds')
    num_beds_loc_entry = float(entry5.get())
    num_beds_loc = (pd.DataFrame(num_beds_loc.loc[num_beds_loc_entry])).head(50)
    labe4.insert(tk.END, num_beds_loc)
    
def zip_code():
    zip_code_loc = data.set_index('ZipCode')
    zip_code_loc_entry = float(entry9.get())
    zip_code_loc = (pd.DataFrame(zip_code_loc.loc[zip_code_loc_entry])).head(50)
    labe4.insert(tk.END, zip_code_loc)


# Created Buttons and connect them to the function
    
button = tk.Button(master= window, text="Result ",width=25, height=5,bg="blue",fg="yellow", command=converter )
button.pack()
button.place(x=100, y=400)

button1 = tk.Button(master= window, text="Go ",width=5, height=2,bg="blue",fg="yellow", command=zip_code )
button1.pack()
button1.place(x=420, y=340)

button2 = tk.Button(master= window, text="Go ",width=5, height=2,bg="blue",fg="yellow", command=num_beds )
button2.pack()
button2.place(x=420, y=240)

button3 = tk.Button(master= window, text="Go ",width=5, height=2,bg="blue",fg="yellow", command=num_Bath )
button3.pack()
button3.place(x=420, y=290)

button4 = tk.Button(master= window, text="Go ",width=5, height=2,bg="blue",fg="yellow", command=price )
button4.pack()
button4.place(x=420, y=170)


# created Text
labe4 = tk.Text(master=window, fg="black",width=80,height=180, font = ('calibre',10,'bold'))
labe4.pack()
labe4.place(x=900, y=20)

window.mainloop()
```
